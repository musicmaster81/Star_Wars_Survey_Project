# Star Wars Survey Project

What with all of the hoopla surrounding Star Wars thanks to the recent release of the long-awaited "Obi-Wan" spinoff series, it would seem appropriate to construct a project around the Star Wars universe. Back in 2015, before the release of "The Force Awakens", the team at FiveThirtyEight conducted a survey wanting to know what fans thought about Episode V: "The Empire Strikes Back." 

The team was able to gather a plethora of responses via a Survey Monkey campaign. This project is dedicated to examining the results of that survey in an attempt to showcase my data cleaning capabilities, arguably the most important component in any Data Scientist's skillset.

### Exploratory Analysis

First, we read the file into a dataframe and examine what it looks like. We also wish to see what are the column names to prepare for cleaning in later steps.

```python
import pandas as pd
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
star_wars.head(10)  # Looking at the first 10 rows of the dataframe
```

```python
star_wars.tail()  # Looking at the last 5 rows of our dataframe
```

```python
# Now, let's take a look at all of the column names. Thirty-eight is a lot.
star_wars.columns
```

### Cleaning the Columns

It is quite obvious before we can begin to perform any meaningful analysis on this data set, we need to get a hold on the column situation. To start off, let's cast the datatypes in the 2 columns containing "Yes/No/NaN" values into booleans. This will make boolean indexing a much easier task

```python
yes_no_map = {
    'Yes': True,
    'No': False
}  # This dictionary defines a mapping from Yes-->True and No-->False
```

```python
# We apply this mapping to the appropriate columns
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no_map)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no_map)
```

```python
# Checking to make sure our changes worked
star_wars.head()
```

## Column Conversions

The next 6 columns represent checkboxes responding to the question, "Which of the following Star Wars films have you seen? Please select all that apply." For example, an respondent who has seen all 6 (we don't count the disgraceful Disney Skywalker saga) episodes will have each episode name inserted into those 6 columns. An individual who has not seen a movie will have a default "NaN" value assigned to that column. 

Instead of having the entire episode name jammed into a single cell, if the cell contains a value, we'll map it to True. If NaN, we'll map it to False and adjust the names of our columns to something more intuitive. 

```python
import numpy as np
# Define our mapping
column_value_map = {
    'Star Wars: Episode I  The Phantom Menace': True,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True,
    np.nan: False
}
```

```python
# We iterate through columns 3-9 and apply our mapping from above
for column in star_wars.columns[3:9]:
    star_wars[column] = star_wars[column].map(column_value_map)
```

```python
# We now change the names of columns 3:9 to something more intuitive
column_name_map = {
    'Which of the following Star Wars films have you seen? Please select all that apply.': 'Seen_Ep_1',
    'Unnamed: 4': 'Seen_Ep_2',
    'Unnamed: 5': 'Seen_Ep_3',
    'Unnamed: 6': 'Seen_Ep_4',
    'Unnamed: 7': 'Seen_Ep_5',
    'Unnamed: 8': 'Seen_Ep_6'
}
```

```python
# We now apply the mapping to our column names
star_wars = star_wars.rename(columns=column_name_map)
```

```python
# Again, let's print the first few values to validate our results
star_wars.head()
```

Now we are making some progress! The next 6 columns ask the respondent to rank the episodes they have seen from most favorite to least favorite. For example, if a respondent really hated episode 1 and loved episode 4, they would place a value of 6 in column 9 and a value of 1 in column 12. We wish to convert these values to floats and rename the columns something more intuitive.  

```python
# Typecast column values as a float
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
# Define our column matching dictionary
episode_rank_column_map = {
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'ep_1_rating',
    'Unnamed: 10': 'ep_2_rating',
    'Unnamed: 11': 'ep_3_rating',
    'Unnamed: 12': 'ep_4_rating',
    'Unnamed: 13': 'ep_5_rating',
    'Unnamed: 14': 'ep_6_rating'
}
```

```python
# Apply the dictionary mapping to the column names
star_wars = star_wars.rename(columns=episode_rank_column_map)
```

```python
# Examine our dataframe to make sure it everything looks good
star_wars.head()
```

## Statistical Analysis

Now that we have our columns in a good place, we are ready to perform some analysis on our data set. We first wish to examine the mean of the ranking columns. We accomplish this by using a dictionary comprehension that uses the column name as the key and the associated column mean as its corresponding value.

```python
# Calculate the mean of each ranking column using dictionary comprehension
ranking_mean = {column: star_wars[column].mean() for column in star_wars.columns[9:15]}
print(ranking_mean)
```

We know wish to perform some exploratory visualizations on these results. We craft a simple bar chart to confirm our findings. 

```python
import matplotlib as plt
%matplotlib inline
average_ranks = pd.Series(data=ranking_mean)
average_ranks.plot.bar()
```

#### Results of Episode Ranking

It would appear that the majority of respondents really enjoyed Episode 5 the most. Recall, a lower ranking is better as a value of "1" is considered an individual's favorite episode.

Personally, I'm a bit upset with this result. Episode 3 is my personal favorite, so seeing that a lot of people did not enjoy it really is a bummer. Perhaps total overall viewership has something to do with such a low rating.  

Recall that we converted several columns into boolean values. We can take the sum of these columns since values of True correspond to 1 and False correspond to 0. This will help us identify how many people have seen each episode.  

```python
# Columns 3-9 are the seen columns. Dictionary Comprehension is used below
total_views = {column: star_wars[column].sum() for column in star_wars.columns[3:9]}
episodes_watched = pd.Series(data=total_views)
```

```python
# Generate a bar graph similar to above to visualize our findings
%matplotlib inline
episodes_watched.plot.bar()
```

#### Results of Total Views for Each Episode

This is now very interesting. First, I for one am shocked that more people have seen Episode 1 than have seen Episode 4, considering that Episode 4 is about 25 years older than Episode 1. Second, it would appear that Episode 3 has the least amount of views. This may explain the discrepancy between Episodes 3 and 5 as Episode 3 will have not had the chance to garner as many positive reviews as Episode 5.

Finally, perhaps the biggest discrepancy between viewer ratings will be found across the gender gap. Let's split the data set up to see what insights can be obtained when we account for gender differences. We can combine this with the results from the "Do you consider yourself to be a fan of the Star Wars film franchise?" column to achieve some real meaningful insights. 

```python
# We use boolean indexing to split our dataframes into male and female exclusive
male_boolean = star_wars['Gender'] == 'Male'
female_boolean = star_wars['Gender'] == 'Female'
male_data = star_wars[male_boolean]
female_data = star_wars[female_boolean]
```

```python
# Let's check to make sure our adjustments worked
male_data['Gender'].value_counts()
```

```python
female_data['Gender'].value_counts()
```

```python
# Let's perform the same analyses as above with our new male/female dataframes
%matplotlib inline
male_ranks = {column: male_data[column].mean() for column in male_data.columns[9:15]}
male_favorites = pd.Series(data=male_ranks)
male_favorites.plot.bar()
```

```python
%matplotlib inline
female_ranks = {column: female_data[column].mean() for column in female_data.columns[9:15]}
female_favorites = pd.Series(data=female_ranks)
female_favorites.plot.bar()
```

#### Gender Delineated Ranking Results

It appears that across both genders, Episode 3 remains the least favorite of the bunch. However, it is interesting to note that females enjoyed Episode 1 substantially more than males. 

```python
# Same analysis as before, only this time to calculate the total viewership
%matplotlib inline
male_views = {column: male_data[column].sum() for column in male_data.columns[3:9]}
total_male_views = pd.Series(data=male_views)
total_male_views.plot.bar()
```

```python
%matplotlib inline
female_views = {column: female_data[column].sum() for column in female_data.columns[3:9]}
total_female_views = pd.Series(data=female_views)
total_female_views.plot.bar()
```

#### Results of Total Viewership across Gender Delineated Data

It would appear that quite a lot more boys have seen each Star Wars episode than girls. However, what is very interesting is that the distribution appears fairly similar as each gender watched episode 5 the most and episode 3 the least.
